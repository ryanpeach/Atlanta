# Walkscore
Now im interested in the walkscore of a place.

```python
from walkscore.api import TransitScore
import pickle
apiKey='ffd1c56f9abcf84872116b4cc2dfcf31'

def maybe_load():
    try:
        with open('transit_score.pickle', 'rb') as handle:
            out = pickle.load(handle)
            print("Loaded!")
            return out
    except:
        print("Did not load.")
        return {}
transit_score = maybe_load()
```

```python
def transitscore(x,y,verbose=False):
    if (x,y) in transit_score:
        return transit_score[(x,y)]["transit_score"]
    else:
        if verbose:
            print("API Call...")

        # reference https://www.walkscore.com/professional/public-transit-api.php
        params = {
            'lat': str(y),
            'lon': str(x),
            'city': 'Atlanta',
            'state': 'GA'
        }
        
        dat = TransitScore(apiKey).makeRequest('score', params)
        transit_score[(x,y)] = dat
        return dat["transit_score"]
    
transitscore(min_x+(max_x-min_x)/10/2, min_y)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
extent = min_x, max_x, min_y, max_y
ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")
ax.set_title("Atlata, Ga")

@interact_manual(alpha=(0.,1.), xbins=(2,20,1), ybins=(2,20,1))    
def overlay(alpha, xbins, ybins):
    ax.imshow(ATL, extent=extent)
    bins = xbins*ybins
    W = np.zeros((int(ybins),int(xbins)))
    for i, j, x, y in ((i,j,x,y) for i, x in enumerate(np.linspace(min_x, max_x, xbins)) for j, y in enumerate(np.linspace(max_y,min_y,ybins))):
        W[j,i] = transitscore(x,y)
        
    ax.imshow(W, extent=extent, alpha=alpha)
    
plt.show()
```

# Combined Histogram

```python


# The Crime Data
cx = np.array(data_map['x']).astype("float")
cy = np.array(data_map['y']).astype("float")

@interact_manual(alpha=(0.,1.), xbins=(2,20,1), ybins=(2,20,1), wc=(0.,3.), wt=(0.,3.), wd=(0.,3.))    
def overlay(alpha, xbins, ybins, wc=1, wt=1, wd=0):
    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)
    extent = min_x, max_x, min_y, max_y
    ax.set_xlabel("Longitude")
    ax.set_ylabel("Latitude")
    img1 = ax.imshow(ATL, extent=extent)
    
    # The transitscore
    T = np.zeros((int(ybins),int(xbins)))
    for i, j, x, y in ((i,j,x,y) for i, x in enumerate(np.linspace(min_x, max_x, xbins)) for j, y in enumerate(np.linspace(max_y,min_y,ybins))):
        T[j,i] = transitscore(x,y)
    
    # The Crime Data
    C, _, _ = np.histogram2d(cx, cy, bins=(xbins,ybins), range=[[min_x, max_x], [min_y, max_y]])
    
    # Distance
    X, Y = np.meshgrid(np.linspace(min_x, max_x, xbins), np.linspace(max_y,min_y,ybins), indexing='xy')
    D = np.sqrt((X-KYS[0])**2. + (Y-KYS[1])**2)
    
    # linear combination of data
    linear  = wt*((T-np.mean(T))/np.std(T))  # Transit score is a positive feature
    linear -= wc*((C-np.mean(C))/np.std(C))  # Crime is a negative feature
    linear -= wd*((D-np.mean(D))/np.std(D))  # Distance is a negative feature
    linear /= np.std(linear)
    img2 = ax.imshow(linear, extent=extent, alpha=alpha)
    fig.colorbar(img2, ax=ax)
    ax.set_title("Atlata, Ga.\nCrime: {}%, Transit: {}%, Distance: {}%".format(wc*100, wt*100, wd*100))
    
    
plt.show()
```
